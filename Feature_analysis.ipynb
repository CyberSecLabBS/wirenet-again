```python
avg_change = 0
for i in range(len(shap_values_tequila_bad[1][0])):
  avg_change = (avg_change*i + (abs(shap_values_tequila_bad[1][0][i]) - abs(shap_values_tequila[1][0][i])))/(i+1)

most_important_differentiated = {}
for i in range(len(shap_values_tequila_bad[1][0])):
  if (abs(shap_values_tequila_bad[1][0][i]) - abs(shap_values_tequila[1][0][i])) > avg_change*100:
    most_important_differentiated[f'{ember_columns[i]}_{i}'] = (abs(shap_values_tequila_bad[1][0][i]) - abs(shap_values_tequila[1][0][i]))

most_important_inverse = {v: k for k,v in most_important_differentiated.items()}
keys = list(most_important_inverse.keys())
keys.sort(reverse=True)

print('ATTRIBUTI DI MAGGIORE IMPORTANZA (ordinati)')
for val in keys:
  print(most_important_inverse[val])
```

```python
most_important = {}
for i in range(len(shap_values_tequila_bad[1][0])):
  if abs(abs(shap_values_tequila_bad[1][0][i]) - abs(shap_values_tequila[1][0][i])) > avg_change*10:  # Diminuito sapendo che molti collassano in uno solo
    if most_important.get(ember_columns[i]):
      most_important[ember_columns[i]] += shap_values_tequila_bad[1][0][i] - shap_values_tequila[1][0][i]
    else:
      most_important[ember_columns[i]] = shap_values_tequila_bad[1][0][i] - shap_values_tequila[1][0][i]

most_important_inverse = {v: k for k,v in most_important.items()}
keys = list(most_important_inverse.keys())
keys.sort()

print('ATTRIBUTI DI MAGGIORE IMPORTANZA (ordinati)')
for val in keys:
  print('{:<60s} \t {}'.format(most_important_inverse[val], val))
```

```python
base_neg = 0
for i in keys:
  if i < 0:
    base_neg += i
  else:
    break
print('SOMMA DEI CONTRIBUTI NEGATIVI RILEVANTI: ', base_neg)
base_pos = 0
for i in keys[::-1]:
  if i > 0:
    base_pos += i
  else:
    break
print('SOMMA DEI CONTRIBUTI POSITIVI RILEVANTI: ', base_pos)
print('Bilancio: ', base_neg+base_pos)
```
